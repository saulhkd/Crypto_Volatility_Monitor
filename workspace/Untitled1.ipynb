```python
from pyspark.sql import SparkSession
import os

spark = SparkSession.builder.appName("CryptoMonitor").getOrCreate()

df_verificacion = spark.read \
    .format("jdbc") \
    .option("url", "jdbc:postgresql://postgres_db:5432/crypto_db") \
    .option("dbtable", "precios_cripto") \
    .option("user", os.getenv("DB_USER")) \
    .option("password", os.getenv("DB_PASSWORD")) \
    .option("driver", "org.postgresql.Driver") \
    .load()

df_verificacion.orderBy(df_verificacion["last_updated"].desc()).show(truncate=False)
print(f"Total guardado: {df_verificacion.count()}")
```
